**Preparacion de librerias**
En este notebook vamos a utilizar las siguientes librerías:
- Scikit-learn y Scipy para el análisis,
- Mtplotlib para los gráficos.
-numpy para el manejo de arrays

```
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs
from numpy import quantile, where, random
import matplotlib.pyplot as plt
```

**preparamaos el dataset** <br>
Crearemos un conjunto de datos de muestra aleatorio para este tutorial usando la función make_blob()

```
random.seed(1)
x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(10,10))
```

**Comprobaremos el conjunto de datos visualizándolo en una gráfica.**

```
plt.scatter(x[:,0], x[:,1])
plt.show()
```

**Definición del modelo y predicción** <br>
Estableceremos el número de estimadores y el valor de contaminación en los argumentos. 

La contaminación define la proporción de valores atípicos en un conjunto de datos.


```
lof = LocalOutlierFactor(n_neighbors=20, contamination=.03)
```

```

#Ajustaremos el modelo con el conjunto de datos x y obtendremos los datos de predicción con el método fit_predict ().
y_pred = lof.fit_predict(x)


#Extraeremos las salidas negativas como valores atípicos.
lofs_index = where(y_pred==-1)
values = x[lofs_index]

#Finalmente, visualizaremos los resultados en un gráfico resaltando las anomalías con un color Rojo
plt.scatter(x[:,0], x[:,1])
plt.scatter(values[:,0],values[:,1], color='r')
plt.show()
```
